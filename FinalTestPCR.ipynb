```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import os
import pickle
```

## read test file from disk

```python
data_test = pd.read_excel('test_set_pcr.xlsx')
 

for col in data_test.columns:
    data_test[col].replace(999,np.nan, inplace=True)

X_test = data_test.drop('ID',axis=1)

```

```python
X_test
```

```python
# Apply robust scaler to image data to scale them and remove outliers
from sklearn.ensemble import VotingClassifier



with open('clf_voting.pkl','rb') as f:
    clf_voting = pickle.load(f)


y_pred = clf_voting.predict(X_test)

y_pred=pd.DataFrame({'pred_pCR':y_pred})
data_pred=pd.merge(y_pred,data_test,left_index=True,right_index=True)
id = data_pred.pop('ID')
data_pred.insert(0,'ID',id)
data_pred
```

```python
## Export final file
data_pred.to_excel('FinalTestPCR.xlsx')
```
